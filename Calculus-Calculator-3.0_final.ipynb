```python
import tkinter as tk
import sympy as sp
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
from sympy.utilities.lambdify import lambdify


class CalculatorGUI:
    def __init__(self, master):
        self.master = master
        master.title("Calculus Calculator")
        
        self.master.configure(bg='#69A6D1')

        # Create a label for the function entry
        self.function_label = tk.Label(master, text="Enter function:",bg = "#94dfff")
        self.function_label.grid(row=0, column=0, padx=5, pady=5)

        # Create a text box for entering the function
        self.function_entry = tk.Entry(master, width=50)
        self.function_entry.grid(row=0, column=1, padx=5, pady=5)

        # Create a label for the differentiation variable entry
        self.diff_var_label = tk.Label(master, text="Differentiation variable:" ,bg = "#c9ebef")
        self.diff_var_label.grid(row=3, column=0, padx=5, pady=5)

        # Create a text box for entering the differentiation variable
        self.diff_var_entry = tk.Entry(master, width=10)
        self.diff_var_entry.grid(row=3, column=1, padx=5, pady=5)

        # Create a differentiation button
        self.diff_button = tk.Button(master, text="Differentiate", command=self.diff , bg = "#94dfff")
        self.diff_button.grid(row=3, column=4, padx=5, pady=5)

        # Create a label for the integration variable entry
        self.intg_var_label = tk.Label(master, text="Integration variable:", bg = "#ffd4b1")
        self.intg_var_label.grid(row=5, column=0, padx=5, pady=5)

        # Create a text box for entering the integration variable
        self.intg_var_entry = tk.Entry(master, width=10)
        self.intg_var_entry.grid(row=5, column=1, padx=5, pady=5)

        # Create a label for the integration limits entry lower
        self.intg_limits_label_1 = tk.Label(master, text="Integration limits lower:", bg = "#fcadb0")
        self.intg_limits_label_1.grid(row=7, column=0, padx=5, pady=5)

        # Create a text box for entering the integration limits
        self.intg_limits_entry_1 = tk.Entry(master, width=20)
        self.intg_limits_entry_1.grid(row=7, column=1, padx=5, pady=5)
        
        # Create a label for the integration limits entry upper
        self.intg_limits_label_2 = tk.Label(master, text="Integration limits upper:" , bg = "#94dfff")
        self.intg_limits_label_2.grid(row=8, column=0, padx=5, pady=5)
        
        # Create a text box for entering the integration limits
        self.intg_limits_entry_2 = tk.Entry(master, width=20)
        self.intg_limits_entry_2.grid(row=8, column=1, padx=5, pady=5)

        # Create an integration button
        self.intg_button = tk.Button(master, text="Integrate",  bg = "#c9ebef", command=self.integrate)
        self.intg_button.grid(row=7, column=4, padx=5, pady=5)
        
         # Create an antiderivative button
        self.intg_antiderivative = tk.Button(master, text="Indefinite Integral",bg = "#ffd4b1" ,command=self.antiderivative)
        self.intg_antiderivative.grid(row=8, column=4, padx=5, pady=5)

        # Create a label for the limit variable entry
        self.lim_var_label = tk.Label(master, text="Limit variable:", bg = "#c9ebef")
        self.lim_var_label.grid(row=10, column=0, padx=5, pady=5)

        # Create a text box for entering the limit variable
        self.lim_var_entry = tk.Entry(master, width=10)
        self.lim_var_entry.grid(row=10, column=1, padx=5, pady=5)

        # Create a label for the limit value entry
        self.lim_val_label = tk.Label(master, text="Limit value:", bg = "#ffd4b1")
        self.lim_val_label.grid(row=11, column=0, padx=5, pady=5)

        # Create a text box for entering the limit
        self.lim_val_entry = tk.Entry(master, width=10)
        self.lim_val_entry.grid(row=11, column=1, padx=5, pady=5)

        # Create a limit button
        self.lim_button = tk.Button(master, text="Evaluate Limit", command=self.limit ,  bg = "#fcadb0")
        self.lim_button.grid(row=10, column=4, padx=5, pady=5)
                
        # Create an extrema_diff  button
        self.extrema_diff = tk.Button(master, text="Extrema_Diff", command=self.find_extrema_diff , bg = "#94dfff")
        self.extrema_diff.grid(row=11, column=4, padx=5, pady =5) 
        
        # Create an extrema_gradient_descent button
        self.extrema_diff = tk.Button(master, text="Extrema_GD", command=self.find_extrema_GD ,  bg = "#c9ebef")
        self.extrema_diff.grid(row=13, column=4, padx=5, pady =5)
        
        # Create a label for the learning rate
        self.learning_rate_label = tk.Label(master, text="Learning rate:", bg = "#fcadb0")
        self.learning_rate_label.grid(row=12, column=0, padx=5, pady=5)
        
        # Create a text box for extrema_gradient_descent learning rate 
        self.learning_rate_entry = tk.Entry(master, width=10)
        self.learning_rate_entry.grid(row=12, column=1, padx=5, pady=5)
        
        # Create a label for initial guess
        self.initial_guess_label = tk.Label(master, text="Initial guess:" ,  bg = "#94dfff")
        self.initial_guess_label.grid(row=13, column=0, padx=5, pady=5)
        
        # Create a text box for extrema_gradient_descent initial guess 
        self.initial_guess_entry = tk.Entry(master, width=10)
        self.initial_guess_entry.grid(row=13, column=1, padx=5, pady=5)
        
        # Create a label for #trial
        self.trial_label = tk.Label(master, text="Number of trial:" , bg = "#c9ebef")
        self.trial_label.grid(row=14, column=0, padx=5, pady=5)
        
        # Create a text box for extrema_gradient_descent #trial
        self.trial_entry = tk.Entry(master, width=10)
        self.trial_entry.grid(row=14, column=1, padx=5, pady=5)
        
        # Create a label for the output
        self.output_label = tk.Label(master, text="")
        self.output_label.grid(row=1, column=1, columnspan=2, padx=5, pady=5)
        self.output_label.config(font = 26)
        
        # Create a button for plotting 
        self.plot_button = tk.Button(master, text="Plot f", command=self.plotting ,  bg = "#ffd4b1")
        self.plot_button.grid(row=14, column=4, padx=5, pady=5)
        
        # Create graph widget
        self.figure = plt.figure(figsize=(4,3), dpi = 100)
        self.canvas = FigureCanvasTkAgg(self.figure, master = master)
        self.canvas.get_tk_widget().grid(row=2, columnspan=5)
        self.canvas.draw()
        
        


    def diff(self):
        function = self.function_entry.get()
        diff_var = self.diff_var_entry.get()

        if function and diff_var:
            f = sp.sympify(function)
            df = sp.diff(f, diff_var)
            self.output_label.config(text=str(df))

    def integrate(self):
        function = self.function_entry.get()
        intg_var = self.intg_var_entry.get()
        intg_limits_lower = self.intg_limits_entry_1.get()
        intg_limits_upper = self.intg_limits_entry_2.get()

        if function and intg_var:
            f = sp.sympify(function)
            a = intg_limits_lower
            b = intg_limits_upper
            integral = sp.integrate(f, (intg_var, a, b))
            self.output_label.config(text=str(integral))

    def limit(self):
        function = self.function_entry.get()
        lim_var = self.lim_var_entry.get()
        lim_val = self.lim_val_entry.get()

        if function and lim_var and lim_val:
            f = sp.sympify(function)
            x = sp.symbols(lim_var)
            val = sp.sympify(lim_val)
            lim = sp.limit(f, x, val)
            self.output_label.config(text=str(lim))

    def antiderivative(self):
        function = self.function_entry.get()
        intg_var = self.intg_var_entry.get()

        if function and intg_var:
            f = sp.sympify(function)
            x = sp.symbols(intg_var)  # use intg_var instead of 'x'
            antiderivative = sp.integrate(f, x)
            self.output_label.config(text=str(antiderivative))
            
    def find_extrema_diff(self):
        function = self.function_entry.get()
        diff_var = self.diff_var_entry.get()

        if function and diff_var:
            f = sp.sympify(function)
            df = sp.diff(f, diff_var)
            ddf = sp.diff(df,diff_var)
            extrema = sp.solve(df, diff_var)
            mini = []
            maxi = []
            for i in extrema: 
                if ddf.subs(diff_var,i)>0:    
                    mini.append(i)
                else:
                    maxi.append(i) 
            self.output_label.config(text= f'Minima is {mini}, Maxima is {maxi}, ')
            
    def find_extrema_GD(self):
        function = self.function_entry.get()
        diff_var = self.diff_var_entry.get()
        alpha = float(self.learning_rate_entry.get())
        guess1 = float(self.initial_guess_entry.get())
        trial = int(self.trial_entry.get())
        
        if function and diff_var:
            f = sp.sympify(function)
            df = sp.diff(f, diff_var)
            for n in range(trial):
                guess1 = guess1 - alpha*df.subs(diff_var, guess1)
            self.output_label.config(text=f'with {trial} trials of shifting x = {guess1} and df(x) = {df.subs(diff_var, guess1)} ')
       
    def plotting(self):
        function = self.function_entry.get()
        diff_var = self.diff_var_entry.get()
        f = sp.simplify(function)
        f1 = lambdify(diff_var, f, "numpy")
        df = sp.diff(f, diff_var)
        x_v = np.linspace(-5, 5, 1000)
        y_v = f1(x_v)
        
        self.figure.clear()
        ax = self.figure.add_subplot(111)
        ax.plot(x_v,y_v)
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        ax.set_title(function)
        self.canvas.draw()
        

    
   
```
